```python
df["acc_total"] = np.sqrt(df["x"]**2 + df["y"]**2 + df["z"]**2)
```

```python
###FEATURE ENGINEERING
## Hard Brake Event
#accel_columns = ['x', 'y', 'z']
#df['total_acceleration'] = (df[accel_columns] ** 2).sum(axis=1) ** 0.5
hard_brake_threshold = 0.987  # Adjust based on your data
df['hard_brake_event'] = df['acc_total'] > hard_brake_threshold

# Sharp Turn Event (Keskin Dönüş)
sharp_turn_threshold = 0.987  # Example threshold
df['event'] = 'Normal'
df.loc[df['acc_total'] > sharp_turn_threshold, 'event'] = 'Sharp Turn'
```

```python
# 1. Ensure timeStamp is in datetime format
df["timeStamp"] = pd.to_datetime(df["timeStamp"])

# 2. Sort by trip
df = df.sort_values(by=["tripID", "timeStamp"]).reset_index(drop=True)

# 3. Calculate elapsed time in seconds for each trip
df["rel_time"] = df.groupby("tripID")["timeStamp"].transform(lambda x: (x - x.min()).dt.total_seconds())

# 4. Assign segment number per 10 seconds
segment_duration = 10  # in seconds
df["insured_no"] = (df["rel_time"] // segment_duration).astype(int)

# 5. Calculate segment start time
segment_start_times = df.groupby(["tripID", "insured_no"])["timeStamp"].min().reset_index()
segment_start_times.rename(columns={"timeStamp": "segment_start_time"}, inplace=True)

# 6. Calculate segment end time
segment_end_times = df.groupby(["tripID", "insured_no"])["timeStamp"].max().reset_index()
segment_end_times.rename(columns={"timeStamp": "segment_end_time"}, inplace=True)

# 7. Statistical summaries per segment (excluding accData, date, time, dtc)
segment_features = df.groupby(["tripID", "insured_no"]).agg({
    "speed": "mean",
    "gps_speed": "mean",
    "battery": "mean",
    "cTemp": "mean",
    "eLoad": "mean",
    "iat": "mean",
    "imap": "mean",
    "kpl": "mean",
    "maf": "mean",
    "rpm": "std",  # variation may be important
    "tAdv": "mean",
    "tPos": "mean",
    "x": "mean",
    "y": "mean",
    "z": "mean",
    "mx": "mean",
    "my": "mean",
    "mz": "mean",
    "acc_total": "mean",
    "hard_brake_event": "sum",  # how many times it occurred in the segment
    "hour": "first",  # for time zone or period info
    "deviceID": "first",
    "day_night": "first",
    "event": "first"
}).reset_index()

# 8. Rename columns for better readability
segment_features.rename(columns={
    "speed": "avg_speed",
    "gps_speed": "avg_gps_speed",
    "battery": "avg_battery",
    "cTemp": "avg_ctemp",
    "eLoad": "avg_eload",
    "iat": "avg_iat",
    "imap": "avg_imap",
    "kpl": "avg_kpl",
    "maf": "avg_maf",
    "rpm": "std_rpm",
    "tAdv": "avg_tadv",
    "tPos": "avg_tpos",
    "x": "avg_x",
    "y": "avg_y",
    "z": "avg_z",
    "mx": "avg_mx",
    "my": "avg_my",
    "mz": "avg_mz",
    "acc_total": "avg_acc_total",
    "hard_brake_event": "count_brake"
}, inplace=True)

# 9. Merge segment start and end times
df_seg = segment_features.merge(segment_start_times, on=["tripID", "insured_no"])
df_seg = df_seg.merge(segment_end_times, on=["tripID", "insured_no"])
```

```python
def label_driver(row):
    is_night = row["day_night"] == "night"

    if is_night:
        if (row["avg_gps_speed"] > 70 and row["avg_acc_total"] > 0.95) or (row["std_rpm"] > 2200):
            return "Aggressive"
        elif (0.92 < row["avg_acc_total"] <= 0.95) or (1800 < row["std_rpm"] <= 2200):
            return "Reckless"
        elif row["avg_gps_speed"] < 40 and row["avg_acc_total"] < 0.8:
            return "Cautious"
        else:
            return "Normal"
    else:
        if (row["avg_gps_speed"] > 80 and row["avg_acc_total"] > 0.98) or (row["std_rpm"] > 2400):
            return "Aggressive"
        elif (0.95 < row["avg_acc_total"] <= 0.98) or (2000 < row["std_rpm"] <= 2400):
            return "Reckless"
        elif row["avg_gps_speed"] < 40 and row["avg_acc_total"] < 0.85:
            return "Cautious"
        else:
            return "Normal"
```

```python
##Analysis of numeric-looking categoricals and categorical-looking numerics

def grab_col_names(dataframe, cat_th=10, car_th=20):
    # cat_cols, cat_but_car
    cat_cols = [col for col in dataframe.columns if dataframe[col].dtypes == "O"]
    num_but_cat = [col for col in dataframe.columns if dataframe[col].nunique() < cat_th and
                   dataframe[col].dtypes != "O"]
    cat_but_car = [col for col in dataframe.columns if dataframe[col].nunique() > car_th and
                   dataframe[col].dtypes == "O"]
    cat_cols = cat_cols + num_but_cat
    cat_cols = [col for col in cat_cols if col not in cat_but_car]

    # num_cols
    num_cols = [col for col in dataframe.columns if dataframe[col].dtypes != "O"]
    num_cols = [col for col in num_cols if col not in num_but_cat]

    # print(f"Observations: {dataframe.shape[0]}")
    # print(f"Variables: {dataframe.shape[1]}")
    # print(f'cat_cols: {len(cat_cols)}')
    # print(f'num_cols: {len(num_cols)}')
    # print(f'cat_but_car: {len(cat_but_car)}')
    # print(f'num_but_cat: {len(num_but_cat)}')
    return cat_cols, num_cols, cat_but_car

cat_cols, num_cols, cat_but_car = grab_col_names(df_seg)
```

```python
## outlier detection
def outlier_thresholds(dataframe, col_name, q1=0.01, q3=0.99):
    q_low = dataframe[col_name].quantile(q1)
    q_high = dataframe[col_name].quantile(q3)
    iqr = q_high - q_low
    lower_bound = q_low - 1.5 * iqr
    upper_bound = q_high + 1.5 * iqr
    return lower_bound, upper_bound

# checks if there are any outliers (returns True/False)
def check_outlier(dataframe, col_name, q1=0.01, q3=0.99):
    low, up = outlier_thresholds(dataframe, col_name, q1, q3)
    return dataframe[(dataframe[col_name] < low) | (dataframe[col_name] > up)].any().any()

# returns the total number of outliers
def grab_outlier_count(dataframe, col_name, q1=0.01, q3=0.99):
    low, up = outlier_thresholds(dataframe, col_name, q1, q3)
    return dataframe[(dataframe[col_name] < low) | (dataframe[col_name] > up)].shape[0]

# replaces outliers with threshold values (lower/upper bounds)
def replace_with_thresholds(dataframe, col_name, q1=0.01, q3=0.99):
    low, up = outlier_thresholds(dataframe, col_name, q1, q3)
    dataframe.loc[dataframe[col_name] < low, col_name] = low
    dataframe.loc[dataframe[col_name] > up, col_name] = up
```

```python
outlier_summary = []

for col in num_cols:
    count = grab_outlier_count(df_seg, col)
    if count > 0:
        outlier_summary.append((col, count))

# DataFrame’e çevir
import pandas as pd
outlier_df_seg = pd.DataFrame(outlier_summary, columns=["Variable", "Outlier Count"]).sort_values(by="Outlier Count", ascending=False)
print(outlier_df_seg)
```

```python
# capping process
# outlier analysis and capping
# recommended q1–q3 values for specific columns
q_dict = {
    "avg_gps_speed": (0.01, 0.99),
    "std_rpm": (0.01, 0.99),
    "avg_my": (0.01, 0.95),
}

# capping based on the recommended q1–q3 ranges
for col, (q1, q3) in q_dict.items():
    if check_outlier(df_seg, col, q1, q3):
        print(f"Capping outliers in {col}... (q1={q1}, q3={q3})")
        replace_with_thresholds(df_seg, col, q1, q3)
```

```python
def one_hot_encoder(dataframe, categorical_cols, drop_first=False):
    dataframe = pd.get_dummies(dataframe, columns=categorical_cols, drop_first=drop_first)
    return dataframe
```

```python
drop_cols = [
    "tripID",                # ID - not meaningful for the model
    "insured_no",            # segment order - sequential numeric but not meaningful
    "segment_start_time",    # timestamp information
    "segment_end_time",      # timestamp information
    "deviceID",              # ID - not encoded, not meaningful for the model
    "avg_dtc",
    "driver_type",
    "avg_iat",
    "driver_type_encoded"    # target variable, should not be in X, will be used as y
]
```

```python
y = df_seg["driver_type_encoded"]
X = df_seg_model
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)

scaler = StandardScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# 1. define the model (with basic settings)
model = LogisticRegression(
    C=1.0,                      # default
    penalty="l2",               # ridge (default)
    solver="liblinear",         # ideal for small datasets and faster training
    multi_class="ovr",          # one-vs-rest
    max_iter=500,
    random_state=42
)

# 2. train the model
model.fit(X_train_scaled, y_train)

# 3. make predictions
y_pred_test = model.predict(X_test_scaled)

# 4. evaluate performance
print("\n---- logistic regression (no grid search) ----")
print("accuracy:", round(accuracy_score(y_test, y_pred_test), 4))
print("precision:", round(precision_score(y_test, y_pred_test, average="macro"), 4))
print("recall:", round(recall_score(y_test, y_pred_test, average="macro"), 4))
print("f1 score:", round(f1_score(y_test, y_pred_test, average="macro"), 4))
print("roc auc:", round(roc_auc_score(y_test, model.predict_proba(X_test_scaled), multi_class="ovr"), 4))
```

```python
y_pred_train = model.predict(X_train_scaled)

print("Train F1:", round(f1_score(y_train, y_pred_train, average="macro"), 4))
print("Test F1 :", round(f1_score(y_test, y_pred_test, average="macro"), 4))
```

```python
from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(model, X_train_scaled, y_train, cv=5, scoring='f1_macro')
print("Cross-Val F1 Mean:", round(cv_scores.mean(), 4))
print("Cross-Val F1 Std :", round(cv_scores.std(), 4))
```

```python
def plot_logistic_feature_importance(model, feature_names, top_n=22):
    import numpy as np
    import pandas as pd
    import matplotlib.pyplot as plt

    if hasattr(model, "coef_"):
        importances = np.abs(model.coef_[0])
    else:
        raise ValueError("Model does not have coef_ attribute")

    feat_imp = pd.Series(importances, index=feature_names).sort_values(ascending=False)
    feat_imp_top = feat_imp.head(top_n)

    print("\nTop Important Features:\n")
    print(feat_imp_top)

    plt.figure(figsize=(10, 6))
    feat_imp_top.plot(kind="barh")
    plt.title("Logistic Regression - Feature Importance")
    plt.xlabel("Coefficient Magnitude")
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()
```
